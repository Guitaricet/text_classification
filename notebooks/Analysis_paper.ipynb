## Ru-Mokoron

### CharCNN

```python
results = pd.read_csv('results/CharCNN_mokoron_noemoji.csv', index_col=0)
results_same = results[results['noise_level_test'] > 0]
results_orig = results[results['noise_level_test'] == -1]

plt.figure(figsize=(8, 5))
plt.title('CharCNN')
plt.xlabel('train noise level')
plt.ylabel('test f1')
plt.plot(results_orig['noise_level_train'], results_orig['f1_test'], color='r', label='test on original')

plot_x = []
plot_y = []

for noise_level in NOISE_LEVELS:
    plot_data = results_same[(results_same.noise_level_train == noise_level) &\
                             (results_same.noise_level_test == noise_level)]
    duplicate = False
    for _, res in plot_data.iterrows():
        plot_x.append(noise_level)
        plot_y.append(res['acc_test'])
        if duplicate:
            print('Duplicate results on noise %s' % noise_level)
        duplicate=True

plt.plot(plot_x, plot_y, label='test on spellchecked')
plt.axhline(0.77)
plt.legend()
```
